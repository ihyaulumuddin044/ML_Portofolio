#Long Short-Term Memory (LSTM) dalam Prediksi Harga Emas
LSTM (Long Short-Term Memory) adalah jenis Recurrent Neural Network (RNN) yang dirancang untuk mengatasi masalah long-term dependencies dalam data sekuensial seperti time series. LSTM sangat cocok untuk prediksi harga emas karena harga emas bersifat waktu ke waktu dan dipengaruhi oleh tren historis.

#6: scaling data

```python
scaler = MinMaxScaler()
scaler.fit(df.Terakhir.values.reshape(-1,1))
```

#7: Restrukturisasi Data


Sliding window adalah teknik dalam analisis data deret waktu di mana kita menggunakan data dari beberapa waktu sebelumnya untuk memprediksi waktu berikutnya.

Bayangkan kita ingin memprediksi harga emas besok berdasarkan harga-harga sebelumnya. Jika kita menggunakan 90 hari terakhir untuk membuat prediksi hari ke-91, maka kita menyebut 90 hari terakhir ini sebagai lebar jendela (window size).

Jadi:

X_train dan X_test berisi daftar harga emas dari 60 hari sebelumnya.
y_train dan y_test berisi harga emas pada hari ke-61 (yaitu harga emas di hari berikutnya setelah 60 hari yang kita gunakan sebagai input).
Dengan cara ini, model dapat belajar pola dari harga-harga sebelumnya dan mencoba memprediksi harga emas di masa depan.

```python
window_size = 90
```

**Training**

```python
train_data = df.Terakhir[:-test_size]
train_data = scaler.transform(train_data.values.reshape(-1,1))
```

```python
X_train = []
y_train = []

for i in range(window_size, len(train_data)):
    X_train.append(train_data[i-window_size:i, 0])
    y_train.append(train_data[i, 0])
```

**Data Testing**

```python
test_data = df.Terakhir[-test_size-window_size:] # Ensure enough data for window
test_data = scaler.transform(test_data.values.reshape(-1,1))
```

```python
X_test = []
y_test = []

for i in range(window_size, len(test_data)):
    X_test.append(test_data[i-window_size:i, 0]) # Use window_size, not 60
    y_test.append(test_data[i, 0])
```

ubah data menjadi numpy

```python
X_train = np.array(X_train)
X_test  = np.array(X_test)
y_train = np.array(y_train)
y_test  = np.array(y_test)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test  = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
y_train = np.reshape(y_train, (-1,1))
y_test  = np.reshape(y_test, (-1,1))
```

```python
print('X_train Shape: ', X_train.shape)
print('y_train Shape: ', y_train.shape)
print('X_test Shape:  ', X_test.shape)
print('y_test Shape:  ', y_test.shape)
```

Membuat model

```python
# def define_model():
#     input1 = Input(shape=(window_size,1))
#     x = LSTM(units = 64, return_sequences=True)(input1)
#     x = Dropout(0.2)(x)
#     x = LSTM(units = 64, return_sequences=True)(x)
#     x = Dropout(0.2)(x)
#     x = LSTM(units = 64)(x)
#     x = Dropout(0.2)(x)
#     x = Dense(32, activation='softmax')(x)
#     dnn_output = Dense(1)(x)

#     model = Model(inputs=input1, outputs=[dnn_output])
#     model.compile(loss='mean_squared_error', optimizer='Nadam')
#     model.summary()

#     return model

def define_model():
    input1 = Input(shape=(window_size, 1))

    x = LSTM(units=128, return_sequences=True)(input1)
    x = Dropout(0.2)(x)

    x = LSTM(units=128, return_sequences=True)(x)
    x = Dropout(0.2)(x)

    x = LSTM(units=64, return_sequences=False)(x)  # Hapus return_sequences=True di LSTM terakhir
    # x = Dense(32, activation='relu')(x)  # Ganti softmax jadi ReLU
    # x = Dense(16, activation='relu')(x)  # Tambahkan satu layer
    dnn_output = Dense(1)(x)  # Output layer tetap 1 neuron

    model = Model(inputs=input1, outputs=[dnn_output])
    model.compile(loss='mean_squared_error', optimizer='adam')  # Gunakan Adam dulu

    model.summary()

    return model
```

```python
model = define_model()
history = model.fit(X_train, y_train, epochs=150, batch_size=32, validation_split=0.1, verbose=1)
```

```python
result = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)
```

```python
MAPE = mean_absolute_percentage_error(y_test, y_pred)
Accuracy = 1 - MAPE
```

```python
print("Test Loss:", result)
print("Test MAPE:", MAPE)
print("Test Accuracy:", Accuracy)
```

```python
y_test_true = scaler.inverse_transform(y_test)
y_test_pred = scaler.inverse_transform(y_pred)
```
